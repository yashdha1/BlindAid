```python
import torchvision
import cv2  
from ultralytics import YOLO 
from torchinfo import summary 
```

```python
model = YOLO("yolov5su.pt") 
summary(model=model)
```

```python
import labelImg
```

```python
vid = 'fellas.mp4'
```

```python
import torch
print("CUDA Available:", torch.cuda.is_available())
print("CUDA Device Count:", torch.cuda.device_count())
print("Current CUDA Device:", torch.cuda.current_device() if torch.cuda.is_available() else "None")
print("CUDA Device Name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "None")
```

```python
# Open webcam
cam = cv2.VideoCapture(vid)

while cam.isOpened():
    det, frame = cam.read()
    
    if det:
        results = model.predict(frame)  # Perform detection
        
        # Draw bounding boxes manually
        for r in results:
            for box in r.boxes.xyxy:
                x1, y1, x2, y2 = map(int, box)
                # Draw rectangle 
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        cv2.imshow("YOLO Detection", frame)  # Show frame with detections

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cam.release()
cv2.destroyAllWindows()
```
